```python
Write a program to accept 2 lists from the user of 5 nos each .Assume each list will have unique nos Now find out how many items 
in these lists are common .
```

```python
# This function takes input and return list of 5 numbers
def user_input():
    l = []
    for i in range(5): 
        l.append(input(f"Enter {i+1} number for the list "))
    return l
        

# This function finds common between two list
def find_common(l1,l2):
    common = []
    if type(l1) and type(l2) == list:
        for i in l1:
            for j in l2:
                if i == j:
                    common.append(i)
    return common
    
list_1 = user_input()
print("first list =",list_1)

list_2 = user_input()
print("second list = ",list_2)

#common list which contains all common numbers
common = find_common(list_1,list_2)
print("common number in both list are as follow: ")

for c in common:
    print(c,end=", ")
```

```python
Write a program to accept 5 unique integers from the user.Make sure if the integer being entered is already present in the list your code displays the message 
“Item already present” and ask the user to reenter the integer. 
```

```python
# function creating list and returning list
def input_():
    lst = []
    for i in range(5):
        t = input(f"Enter {i+1} number for the list")
        while t in lst:  # checking for if number entered is present or not
            print('Item {t} already present'.format(t=t))
            t = input(f"Enter {i+1} number for the list")
        lst.append(t)
    return lst
lst = input_()
print("\nList = ",lst)
```
